```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
import os 
import tarfile
from six.moves import urllib
```

```python
def load_csv(path):
    csv_path = os.path.join(path)
    return pd.read_csv(csv_path)
```

```python
weather_data = load_csv("E:\Labs\AI\Datasets\weatheraus.csv")
```

```python
weather_data.describe(include="all")
```

```python
weather_data.dtypes
```

```python
cat_factors = ["Location", "WindGustDir", "WindDir9am", "WindDir3pm"]
bool_factors = ["RainToday", "RainTomorrow"]
date_factors = ["Date"]
```

```python
for field in bool_factors:
    weather_data[field] = weather_data[field].apply(lambda x : 0. if x == "Yes" else (1. if x == "No" else np.nan))
weather_data
```

```python
import time
from datetime import datetime
for field in date_factors:
    weather_data[field] = weather_data[field].apply(lambda x : time.mktime(datetime.strptime(x, "%Y-%m-%d").timetuple()))
weather_data
```

```python
from sklearn.preprocessing import LabelBinarizer
for field in cat_factors:
    encoder = LabelBinarizer()
    hot_cat = encoder.fit_transform(weather_data[field].apply(lambda x : field + '=' + ("Unk" if type(x) != str else x))).astype(float)
    tmp = pd.DataFrame(hot_cat, columns=encoder.classes_)
    weather_data = pd.concat([weather_data, tmp], axis=1)
    weather_data.pop(field)
weather_data
```

```python
targets = weather_data["RainTomorrow"]
weather_data = weather_data.drop("RainTomorrow", axis='columns')
bool_factors.remove("RainTomorrow")
```

```python
min_corr = 0.1
weather_data.corrwith(targets).apply(lambda x : (abs(x), x)).sort_values(ascending=False).apply(lambda x : x[1]).where(lambda x : abs(x) > min_corr).dropna()
```

```python
def print_nans_info(data):
    nans = data.apply(lambda x : x.apply(lambda x : np.isnan(x).astype(int))).agg('sum').sort_values(ascending=False)
    print("total rows:", len(data))
    for field in nans.index:
        if nans[field] == 0:
            break
        print(field, "nans count:", nans[field])
```

```python
def get_nans_fields(data, ignore_factors=[]):
    nans = data.drop(ignore_factors, axis=1).apply(lambda x : x.apply(lambda x : np.isnan(x).astype(int))).agg('sum')
    return nans.sort_values(ascending=False).where(lambda x : x > 0).dropna()
```

```python
print_nans_info(weather_data)
```

```python
def get_valid_pool(data, field):
    not_nan_axis = data.apply(lambda x : True in x.apply(lambda y : np.isnan(y)).values, axis=1).where(lambda x : x == False).dropna().index
    field_nan_axis = data.apply(lambda x : np.isnan(x[field]), axis=1).where(lambda x : x).dropna().index
    return not_nan_axis, field_nan_axis
```

```python
from sklearn.linear_model import LinearRegression

def train_lr_model(learn, field):
    model = LinearRegression()
    features = learn.drop(field, axis='columns')
    res = learn[field]
    model.fit(features, res)
    return model
```

```python
from sklearn.impute import SimpleImputer

def apply_model(data, field, model, learn_ind, apply_ind):
    imputer = SimpleImputer(strategy="median")
    
    apply = data.drop(field, axis='columns')
    
    imputer.fit(apply)
    apply = imputer.transform(apply)
    res = model.predict(apply).astype(float)

    for i in apply_ind:
        data[field][i] = res[i]
    
    return data
```

```python
def repair_float_fields(data, nan_fields):
    for field in nan_fields.index:
        learn_ind, apply_ind = get_valid_pool(data, field)
        model = train_lr_model(data.loc[learn_ind], field)
        data = apply_model(data, field, model, learn_ind, apply_ind)
        print(field, "\tready!")
    return data
```

```python
nan_fields = get_nans_fields(weather_data, bool_factors)
print(nan_fields)
weather_data = repair_float_fields(weather_data, nan_fields)
weather_data
```

```python
print_nans_info(weather_data)
```

```python
from sklearn.ensemble import RandomForestClassifier

def train_rf_model(learn, field):
    model = RandomForestClassifier(n_estimators=500, max_depth=6, n_jobs=-1)
    features = learn.drop(field, axis='columns')
    res = learn[field].astype(bool)
    model.fit(features, res)
    return model
```

```python
def repair_bool_fields(data, nan_fields):
    for field in nan_fields.index:
        learn_ind, apply_ind = get_valid_pool(data, field)
        model = train_rf_model(data.loc[learn_ind], field)
        data = apply_model(data, field, model, learn_ind, apply_ind)
        print(field, "ready!")
    return data
```

```python
nan_fields = get_nans_fields(weather_data)
print(nan_fields)
weather_data = repair_bool_fields(weather_data, nan_fields)
weather_data
```
